```python
import pandas as pd
import requests
import json
url = "https://api.exchangeratesapi.io/latest?symbols=USD,GBP"
response = requests.get(url)
data = response.text
parsed = json.loads(data)
print(json.dumps(parsed, indent=4))
dfff=pd.DataFrame(parsed)
dfff.insert(0,'currency', dfff.index, True)
dfff.reset_index(drop=True,inplace=True)
display(dfff)
```
